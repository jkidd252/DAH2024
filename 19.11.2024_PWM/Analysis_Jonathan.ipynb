```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
One = (np.loadtxt('data_set_PWM_1.txt')).T
Two = np.loadtxt('data_set_PWM_2.txt').T
Three = np.loadtxt('data_set_PWM_3.txt').T
Four = np.loadtxt('data_set_PWM_4.txt').T
```

```python
print(len(One[:,0]))
```

```python
Two[:,0] = Two[:,0] + One[:,0][-1] + 50
exp = np.concatenate((One, Two), axis=0)
```

```python
plt.plot(One[:,0], One[:,2], label='Sensor 1')
plt.plot(One[:,0], One[:,1], label='Sensor 2')
plt.axhline(19, linestyle='--', label='Set Point = 19')
plt.xlabel('Time [Seconds]')
plt.ylabel(u'Temperature []')
plt.ylim(18, 21)
plt.legend()
plt.show()

plt.plot(Two[:,0], Two[:,2], label='Sensor 1')
plt.plot(Two[:,0], Two[:,1], label='Sensor 2')
plt.axhline(19, linestyle='--', label='Set Point = 19')
plt.xlabel('Time [Seconds]')
plt.ylabel(u'Temperature []')
plt.ylim(18, 20)
plt.legend()
plt.show()

plt.plot(exp[:,0], exp[:,2], label='Sensor 1', zorder=1)
plt.plot(exp[:,0], exp[:,1], label='Sensor 2', zorder=1)
plt.axhline(19, linestyle='--', label='Set Point = 19')
plt.axvspan(One[:,0][-1], Two[:,0][0], alpha=1, color='Gray', label='Stitching Point', zorder=2)
plt.xlabel('Time [Seconds]')
plt.ylabel(u'Temperature []')
plt.ylim(18, 20)
plt.legend()
plt.savefig('Graphs_Report_Jonathan/DC1_plus_2.png')
plt.show()

plt.plot(Three[:,0], Three[:,2], label='Sensor 1')
plt.plot(Three[:,0], Three[:,1], label='Sensor 2')
plt.axhline(19, linestyle='--', label='Set Point = 19')
plt.xlabel('Time [Seconds]')
plt.ylabel(u'Temperature []')
plt.ylim(18, 20.5)
plt.legend()
plt.savefig('Graphs_Report_Jonathan/DC3.png')
plt.show()

plt.plot(Four[:,0], Four[:,2], label='Sensor 1')
plt.plot(Four[:,0], Four[:,1], label='Sensor 2')
plt.axhline(19, linestyle='--', label='Set Point = 19')
plt.xlabel('Time [Seconds]')
plt.ylabel(u'Temperature []')
plt.ylim(18, 21)
plt.legend()
plt.savefig('Graphs_Report_Jonathan/DC4.png')
plt.show()

plt.plot(exp[:,0], (exp[:,1]-exp[:,2]), label='Diff Sensor 2 - Sensor 1 ; Average = '+str(np.round(np.mean(exp[:,1]-exp[:,2]), 2))+' +/- '+str(np.round(np.std(exp[:,1]-exp[:,2]), 2)))
plt.plot(Three[:,0], (Three[:,1]-Three[:,2]), label='Diff Sensor 2 - Sensor 1 ; Average = '+str(np.round(np.mean(Three[:,1]-Three[:,2]),2))+' +/- '+str(np.round(np.std(Three[:,1]-Three[:,2]), 2)))
plt.plot(Four[:,0], (Four[:,1]-Four[:,2]), label='Diff Sensor 2 - Sensor 1 ; Average = '+str(np.round(np.mean(Four[:,1]-Four[:,2]), 2))+' +/- '+str(np.round(np.std(Four[:,1]-Four[:,2]), 2)))
plt.xlabel('Time [Seconds]')
plt.ylabel(u'Temperature []')
plt.ylim(0, 0.6)
plt.legend()
plt.savefig('Graphs_Report_Jonathan/DC_diff.png')
plt.show()
```
